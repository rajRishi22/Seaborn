```python
# 1.Ans
# Bar Plot-Used to represent Numerical value associated with any data type.
# , Scatter Plot-To represent data in form of points.
# , Jointplot-Creates Histogram of both the axes along with their scatter plot.

# ,CatPlot- Represents Data distribution with respect to specific groups

# and Pairplot-Takes all the numerical data pairs and then plots them.
```

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
sns.scatterplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
titanic=sns.load_dataset('titanic')
```

```python
flights=sns.load_dataset('flights')
```

```python
flights
```

```python
flights.groupby('year').passengers.value_counts()
```

```python
flight_heat=(flights.groupby('year').passengers.value_counts().unstack().fillna(0))
```

```python
sns.heatmap(flight_heat)
```
